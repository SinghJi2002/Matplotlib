# Approach 2: Interacting With HTML Backend And Aritistic Layer

```python
#Lets first import a new backend 
from matplotlib.backends.backend_agg import FigureCanvasAgg
#Now we need a container that will store the figure we want print. For that we import this.
from matplotlib.figure import Figure
#Lets create the figure object. Note that this object will actually store the graphics
fig=Figure()
#Lets associate our object/Container with backend
canvas=FigureCanvasAgg(fig)
#Now lets add a subplot to the current figure object.
ax=fig.add_subplot(111)
#Let decode this piece of code, to avoid future confusions
#.add_subplot(111): 
#This method adds a subplot to the figure object.The argument 111 specifies the configuration of the subplot. In this case, 111 
#represents a 1x1 grid, meaning there is only one subplot in the figure. The three-digit number 111 can also be written as 
#(1, 1, 1).
#Lets also discuss about the significance of the variable ax
#By assigning ax to the subplot object, you can use ax to directly manipulate or customize the properties and elements of that 
#specific subplot. It provides a convenient way to access and modify the subplot attributes. For example, you can use ax to set 
#the title of the subplot, customize the axis labels, add annotations, or apply various plotting functions specific to that 
#subplot. Using ax as a reference to the subplot allows you to have more control and flexibility when working with multiple 
#subplots or when applying specific modifications to individual subplots within a figure figure object.

#Now lets access the ax subplot and generate a figure out of it.
ax.plot(3,2,'x')

#Now not that we cannot print the figure directly here, as we are not using Jupyter Notebooks as our backend, but a different one
#that is dependent on HTML. So for now we are just going to save the print in a .png file.
canvas.print_png('test.png')
```

# gca, gcf and get_children

```python
#Lets create a new figure container.
plt.figure()
#Lets plot a point on this new figure
plt.plot(3,2,'o')
#Lets create the reference of the current axes/figure/graph/chart in a variable
ax=plt.gca() #The figure is this cell of the notebook is saved here.
#Yha tk bhi likhne se kaam hojayga
#Though lets set the X-axis and Y-axis lenghts for the chart
ax.axis([0,10,0,10])
#The general for of this command is [Lower Limt X,Upper Limit X, Lower Limit Y, Upper Limit Y]
```

```python
#Lets do it without using the gca
plt.figure()
plt.plot(1,2,'x')
plt.plot(1,3,'o')
```

```python
#Though note that always creating the figure object becomes neccesary only when we want to interact with the artistic layer
#else for scripting interactions direct executions of commands are enough.
plt.plot(1,3,'o')
plt.plot(1,7,'x')
```
