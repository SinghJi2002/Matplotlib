# De-Junkifying A Graph-Applying Tufte's Principles

```python
#Lets first import our libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lets ready our dataset
x_axis=np.array(["Python","SQL","C++","Java","Javascript"])
y_axis=np.array([56, 39, 34, 34, 29])
```

Lets now that we have the graph, we will start implementing Tufte priciples and de-junkify the graph.

### Lets first clear those X and Y axes.

```python
plt.figure(figsize=(10,8))
plt.bar(x_axis,y_axis,align='center')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("The popularity of the languages")
#Lets look at the artists and see what we need to access to implement our plans
ax=plt.gca()
ax.get_children()
```

spines refer to the lines that we want to remove. Lets remove them.

```python
plt.figure(figsize=(10,8))
plt.bar(x_axis,y_axis,align='center')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("The popularity of the languages")
ax=plt.gca().spines.values()
```

```python
plt.figure(figsize=(10,8))
plt.bar(x_axis,y_axis,align='center')
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("The popularity of the languages")
ax=plt.gca().spines
for i in ax.values():
    i.set_visible(False)
```

### Lets highlight One Of the Bars and color rest of them grey.

```python
plt.figure(figsize=(10,8))
positionalArray=np.arange(len(x_axis))
bar=plt.bar(positionalArray,y_axis,align='center',color='grey',alpha=0.6)
bar[0].set_color('blue')
ax=plt.gca().spines
for i in ax.values():
    i.set_visible(False)
plt.xticks(positionalArray,x_axis)
plt.xlabel("Languages")
plt.ylabel("Popularity%")
plt.title("Popularity Of Languages")
```

### Lets remove the y-axis numbers and plant them on the bar

```python
#Lets first clear the y axis
plt.figure(figsize=(10,8))
positionalArray=np.arange(len(x_axis))
bar=plt.bar(positionalArray,y_axis,align='center',color='grey',alpha=0.6)
bar[0].set_color('blue')
ax=plt.gca().spines
for i in ax.values():
    i.set_visible(False)
plt.xticks(positionalArray,x_axis)
plt.yticks([])
plt.xlabel("Languages")
plt.ylabel("Popularity%")
plt.title("Popularity Of Languages")
```

```python
#Lets first clear the y axis
plt.figure(figsize=(10,8))
positionalArray=np.arange(len(x_axis))
bar=plt.bar(positionalArray,y_axis,align='center',color='grey',alpha=0.6)
bar[0].set_color('blue')
ax=plt.gca().spines
for i in ax.values():
    i.set_visible(False)
plt.xticks(positionalArray,x_axis)
plt.yticks([])
plt.xlabel("Languages")
plt.ylabel("Popularity%")
plt.title("Popularity Of Languages")

for i in bar:
    height=i.get_height()
    plt.gca().text(i.get_x() + i.get_width() / 2, i.get_height() - 5, str(int(height)) + '%',
                   ha='center', color='w', fontsize=11)
```

```python
#A simpler way to this is
#Lets first clear the y axis
plt.figure(figsize=(10,8))
positionalArray=np.arange(len(x_axis))
bar=plt.bar(positionalArray,y_axis,align='center',color='grey',alpha=0.6)
bar[0].set_color('blue')
ax=plt.gca().spines
for i in ax.values():
    i.set_visible(False)
plt.xticks(positionalArray,x_axis)
plt.yticks([])
plt.xlabel("Languages")
plt.ylabel("Popularity%")
plt.title("Popularity Of Languages")
plt.bar_label(bar,fmt="%d%%",fontsize=11)
```
