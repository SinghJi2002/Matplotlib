# Box Graphs

```python
#Lets first take the libraries here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lets create databases
data1=np.random.random(size=1000)
data2=np.random.gamma(2,size=1000)
data3=np.random.normal(loc=0,scale=1,size=1000)
```

```python
#Lets create a dataframe out of them.
DataFrame=pd.DataFrame({"Random Data":data1,"Gamma Data":data2,"Normal Data":data3})
```

```python
DataFrame
```

```python
#Lets see some states associated to out dataframe
DataFrame.describe()
```

```python
#Lets make a Box Graph out of the data we already have.
plt.boxplot(DataFrame["Random Data"]);
```

```python
#Lets plot multiple columns
plt.boxplot([DataFrame["Random Data"],DataFrame["Normal Data"],DataFrame["Gamma Data"]]);
```

```python
#Lets just remove those outliers
plt.boxplot([DataFrame["Random Data"],DataFrame["Normal Data"],DataFrame["Gamma Data"]],whis=[0,100]);
```

# inset_axes

```python
plt.figure(figsize=(10,10))
plt.boxplot([DataFrame["Normal Data"],DataFrame["Random Data"],DataFrame["Gamma Data"]])
ax2=plt.gca().inset_axes([0,.6,0.6,0.4])
ax2.hist(np.random.normal(loc=0,scale=1,size=1000),bins=100)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.boxplot([DataFrame["Normal Data"],DataFrame["Random Data"],DataFrame["Gamma Data"]])
ax2=plt.gca().inset_axes([0,.6,0.6,0.4])
ax2.hist(np.random.normal(loc=0,scale=1,size=1000),bins=100)
#Since the axes are collinding we will simply invert them
ax2.yaxis.tick_right()
plt.show()
```
