# General:Line Plots

```python
#Lets first import the important libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Let see the data sets
y_axis=np.array([1,2,3,4,5,6])
y_axis_2=y_axis**2
```

```python
plt.plot?
```

```python
#Lets see the plots 
plt.plot(y_axis,'-o',y_axis_2,'-o')
```

```python
#Lets now make it in a completely new manner. In a detailed manner.
plt.plot(y_axis,'-o',y_axis_2,'-o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Charts")
plt.legend(["Linear Line","Quadratic Lines"])
```

```python
plt.plot(y_axis,'-o',y_axis_2,'-o',[1.5,2.5,3.5],'--r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Charts")
plt.legend(["Linear Line","Quadratic Lines"])
```

# Filled Line Plots

```python
plt.plot(y_axis,'-o',y_axis_2,'-o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Demo Charts")
plt.legend(["Linear Line","Quadratic Lines"])
#This code below fill colour between 2 graphs. Alpha here adjusts thickness of color.
plt.gca().fill_between(range(len(y_axis)),y_axis,y_axis_2,facecolor='blue',alpha=0.25)
```

# Data-Time Based Line Plots

```python
#Lets now start plotting dates.
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
#Lets plot data
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
```

# Playing With Artist Objects

```python
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
ax=plt.gca()
ax.get_children()
#This get_children command is run in reference to X axis. This is just to demostrate how each and every label here functions as
#an artist itself.
```

```python
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
ax=plt.gca().xaxis
ax.get_children()
#This get_children command is run in reference to X axis. This is just to demostrate how each and every label here functions as
#an artist itself.
```

```python
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
ax=plt.gca().xaxis
#The code above creates an access to artists of x axis.
for i in ax.get_ticklabels():
    i.set_rotation(45)
```

As observable how we accessed each and every label with help of tickers and then rotated them all using the code above.

# Lets now make a Complete Graph

```python
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
ax=plt.gca().xaxis
#The code above creates an access to artists of x axis.
for i in ax.get_ticklabels():
    i.set_rotation(45)
plt.xlabel("Dates")
plt.ylabel("Units")
plt.legend(["Linear","Squared"])
plt.title("Linear Progression v/s Quadratic Propagation")
```

# LaTeX

```python
data_array=np.arange('2007-01-01','2007-01-07',dtype='datetime64[D]')
plt.plot(data_array,y_axis,'-o',data_array,y_axis_2,'-o')
ax=plt.gca().xaxis
#The code above creates an access to artists of x axis.
for i in ax.get_ticklabels():
    i.set_rotation(45)
plt.xlabel("Dates")
plt.ylabel("Units")
plt.legend(["Linear","Squared"])
#Look How title looks different
plt.title("Exponential ($x^2$) vs. Linear ($x$) performance")
#The usage of $ and ^ is new and comes under LaTeX.
```
