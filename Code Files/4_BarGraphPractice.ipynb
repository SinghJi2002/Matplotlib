# Bar Charts: Normal Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lets declare datasets
x_axis=np.array([0,1,2,3,4,5,6])
y_linear_data=np.array([1,2,3,4,5,6,7])
y_quadratic_data=y_linear_data**2
```

```python
plt.bar?
```

```python
#Lets see the first plot
#Single Plot
plt.bar(x_axis,y_linear_data,width=0.3,color='green')
```

# MultiCharts 

```python
#Multiplot Bar Graphs
plt.bar(x_axis,y_linear_data,width=0.3,color='green')
plt.bar(x_axis,y_quadratic_data,width=0.3,color='red')
```

As observable here we could now have the results we wanted. Its because we see that over the same point the two graphs are being pointed as a result one graph overshadows the other. Lets see another approach to form it.

```python
new_x=[]
for i in x_axis:
    new_x.append(i+0.3)#+0.3 is being used because width of the graph is 0.3
new_x=np.array(new_x)
plt.bar(x_axis,y_linear_data,width=0.3,color='green')
plt.bar(new_x,y_quadratic_data,width=0.3,color='red')
```

Now you see it working just fine.

# Error Charts:yerr

```python
from random import randint
linear_err = [randint(1,4) for x in range(len(y_linear_data))] 
plt.bar(x_axis,y_linear_data,width=0.3,yerr=linear_err,color='green')
```

# Stacked Plotting: bottom

```python
plt.bar(x_axis,y_linear_data,width=0.3,color='green')
plt.bar(x_axis,y_quadratic_data,width=0.3,bottom=y_linear_data,color='red')
#Its kind of multiploting but we see that x values this time are the same.
```

# Staked Horizontal Graphs:barh and left and height

```python
#The meaning of x and y axis do not change, here although the ploting is horizontal. The x-axis and y-axis arrays are same in 
#both case.
plt.barh(x_axis,y_linear_data,height=0.3,color='green')
plt.barh(x_axis,y_quadratic_data,height=0.3,left=y_linear_data,color='red')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace this with your own data)
data = np.random.randn(10)  # 10 data points for demonstration

# Calculate mean and 95% confidence interval
mean = np.mean(data)
print(mean)
confidence_interval = 1.96 * np.std(data) / np.sqrt(len(data))

# Create the bar chart with error bars representing the confidence interval
plt.bar(range(len(data)), data, yerr=confidence_interval, capsize=5)

# Set plot labels and title
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Bar Chart with 95% Confidence Interval for the Mean')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
values = [10, 20, 15]
threshold = 15

# Create the bar graph
fig, ax = plt.subplots()
bars = ax.bar(categories, values)

# Access and modify each bar one by one
for i, bar in enumerate(bars):
    if values[i] > threshold:
        bar.set_color('red')

# Show the plot
plt.show()
```
