# Seaborn

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Lets create databases that we are going to plot
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
#Lets begin the plotting process
plt.figure()

plt.hist(v1,alpha=0.5,bins=np.arange(-50,150,5),label='v1')
plt.hist(v2,alpha=0.5,bins=np.arange(-50,150,5),label='v2')
plt.legend()
```

```python
#Well that went quite will. Next thing is we add some seaborn utility
#Lets begin the plotting process
plt.figure()

plt.hist(v1,alpha=0.5,bins=np.arange(-50,150,5),label='v1')
plt.hist(v2,alpha=0.5,bins=np.arange(-50,150,5),label='v2')
plt.legend()
sns.set()
#The sns.set() command helps us set some predefined designs for our plot, without you know making much changes ourselves via
#multiple commands. The evolved version is sns.set_style("Your choice").
```

```python
sns.set?
```

```python
#Let work on another wonder. Lets create a new type of hist. Here we see in the old one that the bars that are colliding are both
#visible. What I want is, only those to be viewed, that are one over the other.
plt.hist([v1,v2],histtype="barstacked",bins=np.arange(-50,150,5));
```

```python
#Let now mix these graphs you know. An then use seaborn to plot a kde.
plt.figure()
plt.hist([v1,v2],histtype="barstacked",density=True)
v3=np.concatenate((v1,v2))
sns.kdeplot(v3)
```

```python
#Another smaller but easier method to this was
sns.histplot(v3,kde=True)
```

Note that the above graph is drawn on v3, not v1 and v2 hence the colour changes

```python
#Now we are going to do joint plots. We have already done this. You may view the histogram section, most importantly, the grid
#plot part. But well It took huge chain of commands for it to be achieved in the first place, so what we are going to do is,
#make it with seaborn, in just few lines.

sns.jointplot(x=v1,y=v2,alpha=0.5)
```

One of the graph is the normal distribution v1, the other is normal distribution v2, while the common plot is scatter plot of v1 and v2

```python
#What if we want some other common plot. Well that can be achieved by the help of the kind attribute. See below
sns.jointplot(x=v1,y=v2,alpha=0.5,kind='hex')
```

```python
#Now there is the space attribute that controls the space between the joint plots. Here is examplar implementation.
sns.jointplot(x=v1,y=v2,alpha=0.5,kind='hex',space=0)
```

```python
#Here is another example
sns.jointplot(x=v1,y=v2,alpha=0.5,kind='kde',space=0)
```
