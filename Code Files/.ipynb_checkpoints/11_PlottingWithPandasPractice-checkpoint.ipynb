# Lets plot, but lets plot using pandas without calling out matplotlib exclusively

```python
#Lets first import all important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lets design the datasets we want to plot

dataframe=pd.DataFrame({'A':np.random.randn(365).cumsum(0),
                       'B':np.random.randn(365).cumsum(0)+20,
                       'C':np.random.randn(365).cumsum(0)-20})
```

```python
#Lets see the dataframe required
dataframe.head()
```

```python
#Lets start ploting the data
#But first we need to see you know all the kinds of style of plots we have. Command is.
plt.style.available
```

```python
#Lets set the style to 'seaborn-colorblind'
plt.style.use('seaborn-colorblind')
```

# dataframe.plot

```python
#Now without calling plt exclusively, lets plot dataframe using the dataframe itself. The general command is simply df.plot()
dataframe.plot();
```

```python
dataframe.plot.hist?
```

```python
#We have just seen a simple line plot here. Well say, what if I want a scatter plot, or a histogram or box plot and etc, well
#there two methods to achieve this. Lets see the codes.

#Method1:
dataframe.plot('A','B',kind='scatter')
```

```python
#Method 2:
dataframe.plot.scatter('A','B')
```

As observable,2 completely similar plot are formed.

Lets add few more attribute, to this and we will see few more examples

```python
dataframe.plot.scatter?
```

```python
dataframe.plot.scatter('A','C',c='B',s=dataframe['B'],colormap='viridis');
#Note in place of passing Database[column] we are simply passing the column name for x and y axes.
```

```python
#Lets see you know a few more examples. Lets see histograms.
dataframe.plot.hist(by=dataframe['A'],bins=365)
```

```python
#As observable an error is generated. The right code is.
dataframe['A'].plot.hist(bins=10)
```

```python
#Lets discuss about aspect ratio. Here is how we apply it use.
ax=dataframe.plot.scatter('A','C',c='B',s=dataframe['B'],colormap='viridis')
ax.set_aspect('equal');
```

```python
#Here is another example, box plots.
dataframe.plot.box()
```

```python
#Another example from the histograms
dataframe.plot.hist(alpha=0.5,bins=100)
```

```python
#Another example. This time we see kernel density estimation charts.
dataframe.plot.kde();
```
