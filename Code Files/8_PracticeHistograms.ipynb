# Lets make Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.normal?
```

```python
#Lets see applications
plt.figure()
#Lets create fig plots
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,sharex=True)
#Lets store the grid reference variables in a list.
axs=[ax1,ax2,ax3,ax4]
#Lets iterate through each grid. And also plot the histograms
for i in range(0,len(axs),1):
    order=10**(i+1)
    sampleSet=np.random.normal(loc=0,scale=1,size=order)
    #Let plot the data
    axs[i].hist(sampleSet)
```

```python
#The general bin setting is 10,lET reset it to 100
plt.figure()
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,sharex=True)
axs=[ax1,ax2,ax3,ax4]
for i in range(0,len(axs),1):
    arange1=10**(i+1)
    sampleSet=np.random.normal(loc=0,scale=1,size=arange1)
    axs[i].hist(sampleSet,bins=100)
```

# Lets see gridspec

```python
#Create the general grid using the gridspec functionality
#But yes. Do import its libraries
import matplotlib.gridspec as gridspec

plt.figure(figsize=(10,10))

GeneralGrid=gridspec.GridSpec(3,3)

#Lets now draw the cut outs

rightMostUpper=plt.subplot(GeneralGrid[0,1:])
leftMostLower=plt.subplot(GeneralGrid[1:,0])
theBiggerOne=plt.subplot(GeneralGrid[1:,1:])
```

```python
#Now lets plot.
plt.figure(figsize=(10,10))
GeneralGrid=gridspec.GridSpec(3,3)
rightMostUpper=plt.subplot(GeneralGrid[0,1:])
leftMostLower=plt.subplot(GeneralGrid[1:,0])
theBiggerOne=plt.subplot(GeneralGrid[1:,1:])

#Lets plot with the leftMost Upper Grid
rightMostUpper.hist(np.random.normal(loc=0,scale=1,size=1000),bins=100)
leftMostLower.hist(np.random.normal(loc=0,scale=1,size=1000),orientation='horizontal',bins=100)
theBiggerOne.scatter(np.random.random(1000),np.random.normal(loc=0,scale=1,size=1000))
```

```python
#Now we do 3 things
#1) Invert the x_axis
#2) Turn density on
#3) Remove the left and right padding

plt.figure(figsize=(10,10))

import matplotlib.gridspec as gridspec

GridGeneral=gridspec.GridSpec(3,3)

leftLowerMost=plt.subplot(GridGeneral[0,1:])

rightUpperMost=plt.subplot(GridGeneral[1:,0])

theBottomPlot=plt.subplot(GridGeneral[1:,1:])

leftLowerMost.hist(np.random.normal(loc=0,scale=1,size=1000),bins=100,density=True)

rightUpperMost.hist(np.random.normal(loc=0,scale=1,size=1000),bins=100,density=True,orientation='horizontal')

rightUpperMost.invert_xaxis()

theBottomPlot.scatter(np.random.random(1000),np.random.normal(loc=0,scale=1,size=1000))
```
